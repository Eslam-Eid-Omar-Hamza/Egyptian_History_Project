This notebook demonstrates how the trained YOLOv8 model detects famous Egyptian landmarks such as the Pyramids, the Sphinx, and Pharaoh statues.  
The goal is to identify landmarks from images and provide historical information about each one.


```python
import torch
from ultralytics import YOLO
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the trained YOLO model
model = YOLO('models/best_giza_landmarks.pt')
print("‚úÖ Model loaded successfully!")
```

```python
# Test a sample image
image_path = 'assets/Pharaoh_Small1.png'  # replace with an actual test image
results = model(image_path)

# Show results
results[0].show()  # Opens a window with bounding boxes
```

```python
# Display the image inside the notebook
result_image = results[0].plot()
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Prediction Result")
plt.show()
```
